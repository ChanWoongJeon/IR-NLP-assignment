1. MOVIE 2012

split word and its frequency

```python
import re
import string
fre = {}
doc_txt = open('2012.txt', 'r')
text_string = doc_txt.read().lower()
match = re.findall(r'\b[a-z]{3,15}\b', text_string)
for word in match:
    count = fre.get(word,0)
    fre[word] = count + 1
print("before remove stopwords",len(fre))
```

remove stopword

```python
fre_list = fre.keys()
lists=[]
for i in range(len(stop_word)):
    for words in fre_list:
        if stop_word[i] == words:
            lists.append(words)
for j in range(len(lists)):
    del fre[lists[j]]
print("After remove stopwords",len(fre))
```

Listed in order of frequency

```python
ffre_sort = sorted(fre.items(), key=lambda x: x[1], reverse=True)
print(ffre_sort)
```

select top 100 frequency(because there are so many words)

```python
aaa={}
for i in range(100):
    aaa[ffre_sort[i][0]]=ffre_sort[i][1]
print("top 100 words\n",aaa)
```

Words distribution(Top 100)

```python
import matplotlib.pyplot
plt.figure(figsize=(20, 5))
plt.bar(aaa.keys(), aaa.values(), color='g')
plt.show()
```

Words distribution(Top 100)(each graph split 20)

```python
import matplotlib.pyplot
for i in range(5):
    aaa1={}
    for j in range(20):
        aaa1[ffre_sort[j+i*20][0]]=ffre_sort[j+i*20][1]
    plt.figure(figsize=(20, 5))
    plt.bar(aaa1.keys(), aaa1.values(), color='g')
    plt.show()
```
